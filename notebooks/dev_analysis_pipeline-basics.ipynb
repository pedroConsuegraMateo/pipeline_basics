### Refactoring

```python
def analysis(df, group, agg, file='../data/result.csv'):
    analyzed = cars.groupby(by=group)[agg].mean()
    analyzed.to_csv(file)
    return analyzed
    
```

```python
analysis(cars, 'Make', 'Combined MPG')
```
